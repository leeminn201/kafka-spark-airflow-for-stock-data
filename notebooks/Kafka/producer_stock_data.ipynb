```python
def get_data_raw(symbols):
    df = pd.DataFrame()
    for symbol in symbols:
        temp = pd.DataFrame(yf.download(symbol, period='3m', interval='1m'))[-1:]
        temp['Symbol'] = symbol
        df = df.append(temp)
    return df
```

```python
def get_price_realtime(symbols, data_historical):
    price_realtime = []
    for i in range(len(symbols)):
        temp = float(data_historical[i].iloc[-1:]['Open'])
        price_realtime.append(temp)

    column_labels = ['Ticker', 'Price']
    price = pd.DataFrame(columns=column_labels)

    price['Price'] = price_realtime
    price['Ticker'] = symbols

    return price
```

```python
while(1):
    df =  get_data_raw(symbols)
    print(df)
    time.sleep(5)
```
